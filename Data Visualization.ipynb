# Task
### Compare the trend of 5 countries

```python
df.sort_values(by='Total',ascending=False,axis=0,inplace=True) #sorting frame
```

```python
df.head()
```

```python
df_top5 = df.head(5)
df_top5 = df_top5[years].transpose()
```

```python
df_top5.plot(kind='line')
plt.title("Immigration trend from top 5 countries")
plt.ylabel("Number of imm")
plt.xlabel("Years")
plt.show()
```

```python
df_top5.plot(kind='area',
            stacked=False, #overlapping stacks allowed if false else one over other only
            figsize=(20,10),
            alpha=0.25 #color transparency control
            )
plt.title("Immigration trend of top 5 countries")
plt.ylabel("Number of imm")
plt.xlabel("Years")
plt.show()
```

```python
#Scripting layer (procedural methof of plotting, above also same)
df_top5.plot(kind='area',figsize=(20,10),alpha=0.35)
plt.title("Immigration trend of top 5 countries")
plt.ylabel("Number of imm")
plt.xlabel("Years")
plt.show()
```

```python
#Artist layer (object oriented method of plotting, useful for multiple plots)
ax = df_top5.plot(kind='area',figsize=(20,10),alpha=0.35)
ax.set_title("Immigration trend of top 5 countries")
ax.set_ylabel("Number of imm")
ax.set_xlabel("Years")
```
